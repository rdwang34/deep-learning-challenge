```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
import tensorflow.keras as keras

import pandas as pd 
application_df = pd.read_csv("charity_data.csv")
application_df.head()
```

```python
application_df.drop(['EIN','NAME'], axis=1)
```

```python
value_count = application_df['APPLICATION_TYPE'].value_counts()
value_count
```

```python
application_types_to_replace = list(value_count[value_count<500].index)

for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

application_df['APPLICATION_TYPE'].value_counts()
```

```python
class_value_count = application_df['CLASSIFICATION'].value_counts()
class_value_count
```

```python
greater_than = application_df['CLASSIFICATION'].value_counts().loc[lambda x : x > 1]
greater_than
```

```python
classifications_to_replace = list(greater_than[greater_than<1000].index)

for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    

application_df['CLASSIFICATION'].value_counts()
```

```python
app_dummies = pd.get_dummies(application_df)
app_dummies.head()
```

```python
X = app_dummies.drop('IS_SUCCESSFUL', axis=1)
y = app_dummies['IS_SUCCESSFUL'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=45)
```

```python
scaler = StandardScaler()

X_scaler = scaler.fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

number_input_features = X_train_scaled.shape[1]
```

```python
def create_model(hp):
    nn_model = tf.keras.models.Sequential()

    activation = hp.Choice('activation',['relu','relu','sigmoid'])
    
    nn_model.add(tf.keras.layers.Dense(units=hp.Int('first_units',
        min_value=1,
        max_value=10,
        step=2), activation=activation, input_dim=number_input_features))

    for i in range(hp.Int('num_layers', 1, 6)):
        nn_model.add(tf.keras.layers.Dense(units=hp.Int('units_' + str(i),
            min_value=1,
            max_value=10,
            step=2),
            activation=activation))
    
    nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

    nn_model.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])
    
    return nn_model
```

```python
import keras_tuner as kt

tuner = kt.Hyperband(
    create_model,
    objective="val_accuracy",
    max_epochs=20,
    hyperband_iterations=2)
```

```python
tuner.search(X_train_scaled,y_train,epochs=20,validation_data=(X_test_scaled,y_test))
```

```python
best_hyper = tuner.get_best_hyperparameters(1)[0]
best_hyper.values
```

```python
best_model = tuner.get_best_models(1)[0]
model_loss, model_accuracy = best_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
best_model.save('AlphabetSoupCharity_Optimization.h5')
```
